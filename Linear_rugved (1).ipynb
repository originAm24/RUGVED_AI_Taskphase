```python
arr=np.array(df.drop(["CO2 Emissions(g/km)"], axis=1))
y=df["CO2 Emissions(g/km)"]
y=np.array(y)

def compute_gradient(X, y, w, b):
  m,n=X.shape
  dj_dw = np.zeros((n,))
  dj_dw = np.array(dj_dw, dtype=np.float64)
  w = np.array(w, dtype=np.float64)
  dj_db=0.
  for i in range(m):
    err=(np.dot(w,X[i])+b)-y[i]
    for j in range(n):
      dj_dw[j]=dj_dw[j]+err*X[i, j]
    dj_db=dj_db+err

  dj_dw=dj_dw/m
  dj_db=dj_db/m
  return dj_dw, dj_db
```

```python
def gradient_descent():
  m,n=arr.shape
  w=np.zeros((n,))
  w = np.array(w, dtype=np.float64)
  b=0
  for i in range(1000):
    dj_dw, dj_db=compute_gradient(arr, y, w, b)
    w=w-0.0003*dj_dw
    b=b-0.0003*dj_db

  return w, b
```

```python
W, B=gradient_descent()
```
